Importing packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Cleaning data by removing duplicates

```python
# Remove any rows with missing values
df.dropna(inplace=True)

# Remove any duplicates
df.drop_duplicates(inplace=True)
```

Data now duplicate free

```python
df = pd.read_csv("StudentsPerformance.csv")
```

# New Section

```python
df.head()
```

checking no of students in each race

```python
direction_counts = df['race/ethnicity'].value_counts()
direction_counts.plot(kind='bar', color='Green')
plt.xlabel('race/ethnicity')
plt.ylabel('count')
plt.title('Bar Chart of race/ethnicity')
plt.show()
plt.savefig('/content/StudentPerformance/results/race.png')
```

calculating average score

```python
df['avgscore'] = df['math score'] + df['reading score'] +df['writing score']
```

Using Histogram to plot average score to understand range of average score

```python
plt.hist(df['avgscore'], bins=20, edgecolor='black')
plt.xlabel('avgscore')
plt.title('Histogram of avgscore')
plt.show()
plt.savefig('/content/StudentPerformance/results/avgshist.png')
```

Using Boxplot comparing average score based on race

```python
import matplotlib.pyplot as plt
sns.boxplot(data=df, x="race/ethnicity", y="avgscore")
plt.title('average Score by Race/Ethnicity')
plt.xlabel("Race/Ethnicity")
plt.ylabel("avgscore")
plt.show()
plt.savefig('/content/StudentPerformance/results/avgsbyrace.png')
```

Using Violinplot understanding relation between parentallevel of education and average score of student

```python

avgs=sns.violinplot(x="parental level of education", y="avgscore", data=df)
img=avgs.get_figure()
img.savefig('/content/StudentPerformance/results/avgs.png')
```

Ploting corelaton Heatmap

```python
cm=sns.heatmap(df.corr(), cmap="coolwarm", annot=True)
plt.title("corelation Matrix")
plt=cm.get_figure()
plt.savefig('/content/StudentPerformance/results/cm.png')
```

Ploting mathscore by race using Pie chart
